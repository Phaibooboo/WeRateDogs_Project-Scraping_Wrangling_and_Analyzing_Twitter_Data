### 1. Your friend works for an antique book shop that sells books between 1800 and 1950 and wants to quickly categorise books by the century and decade that they were written. Write a program that takes a year (e.g. 1872) and outputs the century and decade (e.g. "Eighteenth Century, Seventies")

```python
"""PYTHON PROJECT TO GROUP BOOKS WRITTEN BETWEEN 1800 AND 1950 INTO CENTURY AND DECADE"""


# Takes a year input and check if it's

def categorize_book():
    first_input = input("What year was the book written? ")
    try:
        user_input = int(first_input)
        if user_input < 1800 or user_input > 1950:
            print("Kindly enter a year between 1800 and 1950")
            categorize_book()
        elif user_input >= 1800 or user_input <= 1950:
            century = ""
            decade = ""
            # converts year in int to string to extract the second digit
            year = str(user_input)
            digit = year[1]
            century_digit = int(digit)
            #  checks century name based on second digit of year
            if century_digit == 9:
                century = "Twentieth Century"
            elif century_digit == 8:
                century = "Nineteenth Century"
            # extracts third digit of year
            digit = year[2]
            decade_digit = int(digit)
            # checks decade based on the third digit of year
            if decade_digit == 0:
                decade = "Noughties"
            elif decade_digit == 10:
                decade = "Teenies"
            elif decade_digit == 2:
                decade = "Twenties"
            elif decade_digit == 3:
                decade = "Thirties"
            elif decade_digit == 4:
                decade = "Forties"
            elif decade_digit == 5:
                decade = "Fifties"
            elif decade_digit == 6:
                decade = "Sixties"
            elif decade_digit == 7:
                decade = "Seventies"
            elif decade_digit == 8:
                decade = "Eighties"
            elif decade_digit == 9:
                decade = "Nineties"

            print(f"{century}, {decade}")
    except ValueError:
        print("You entered an invalid input. Please enter a numerical value for the year between 1800 and 1950 e.g 1930")
        categorize_book()


categorize_book()
```

### 2. Write a program that simulates a lottery. 

**The program should have a list of seven numbers that represent a lottery ticket.
 It should then generate seven random numbers.
 After comparing the two sets of numbers, the program should output a prize based on the number of matches:
·         £20 for three matching numbers
·         £40 for four matching numbers
·         £100 for five matching numbers
·         £10000 for six matching numbers
·         £1000000 for seven matching numbers**


```python
#Case 1: Here, the company generates 7 random numbers for each client. The client is allowed to choose his/her 7 winning numbers.

"""WELCOME TO MY LOTTERY PLACE. 7 NUMBERS RANDOMLY SELECTED FROM NUMBERS 0 TO 9 WILL BE YOUR LUCKY LOTTERY NUMBERS. BEST OF LUCK WINNING THAT GRAND PRIZE."""



from numpy import random

#Generate seven random numbers which determine winnings
x =random.randint(10, size=(7))

#The customer chooses his/her 7 winning lottery numbers
while True:    
    lottery_numbers = input('Enter 7 numbers between 0 and 9: ')

    if len(lottery_numbers) != 7:
        print('please, enter 7 numbers')
        continue
    if lottery_numbers.isnumeric() == False:
        print('please, enter valid digits only')
        continue
    break

#append any winning lottery number to empty list so its length can be counted
lot_num = []

for i in lottery_numbers:
    i = int(i)
    lot_num.append(i)
    
match = []
for i in lot_num:
    if i in x:
        match.append(i)
        
if len(match) == 3:
    print('you have won £20')
elif len(match) == 4:
    print('you have won £40')
elif len(match) == 5:
    print('you have won £100')
elif len(match) == 6:
    print('you have won £10000')
elif len(match) == 7:
    print('you have won £1000000')
else:
    print('Try again in 6 hours')
    
```

```python
#Case 2 The winning numbers are fixed. The company randomly generates 7 numbers for each client(casino style)
"""WELCOME TO MY LOTTERY PLACE. THE LUCKY LOTTERY NUMBERS ARE SHOWN BELOW, WHILE THE RANDOM NUMBERS WILL BE RANDOMLY SELECTED FROM NUMBERS 1 TO 20. BEST OF LUCK WINNING THAT GRAND PRIZE."""



lottery_numbers = [1, 3, 7, 8, 13, 19, 16]
random_numbers = []
matching_numbers = 0

# Starting the lottery and creating the list of random numbers
print("Welcome to Wofai's Lottery number game")
while True:
    user_input = input("Enter the word 'random' to get a list of 7 random numbers: ")
    if user_input == "random":
        while len(random_numbers) < 7:
            random_num = random.randint(1, 21)
            if random_num not in random_numbers:
                random_numbers.append(random_num)
        print(f"Your random numbers are {random_numbers}. The secret lottery numbers are {lottery_numbers}")
    else:
        print("Please enter a valid input")
        continue
    break 


# checking for matching numbers
for lottery_num in lottery_numbers:
    for random_num in random_numbers:
        if random_num == lottery_num:
            matching_numbers = matching_numbers + 1

print(f"You got {matching_numbers} matching number(s)")

# Checking if the user eligible for a prize based on the amount of matching numbers
if matching_numbers == 3:
    print("You have won £20 for getting three matching numbers!")
elif matching_numbers == 4:
    print("You have won £40 for getting four matching numbers!")
elif matching_numbers == 5:
    print("You have won £100 for getting five matching numbers!")
elif matching_numbers == 6:
    print("You have won £10000 for getting six matching numbers!")
elif matching_numbers == 7:
    print("You have won £1000000 for getting seven matching numbers!")
else:
    print("Oops! You did not get enough matching numbers to win a prize. Try again in 6 hours.")
```

# 3. Exception Handling:
    
**Simple ATM program
Using exception handling code blocks such as try/ except / else / finally, write a program that simulates an ATM machine to withdraw money.
(NB: the more code blocks the better, but try to use at least two key words e.g. try/except)**

Tasks:
1.       Prompt user for a pin code
2.       If the pin code is correct then proceed to the next step, otherwise ask a user to type in a password again. You can give a user a maximum of 3 attempts and then exit a program.
3.       Set account balance to 100.
4.       Now we need to simulate cash withdrawal
5.       Accept the withdrawal amount
6.       Subtract the amount from the account balance and display the remaining balance (NOTE! The balance cannot be negative!)
7.       However, when a user asks to ‘withdraw’ more money than they have on their account, then you need to raise an error an exit the program.


```python
"""WELCOME TO MY ATM PROJECT. KINDLY NOTE THAT THE DEFAULT ATM PIN TO TEST THIS PROGRAM IS 1234. ANYTHING OTHER
THAN THAT WILL RETURN AN INCORRECT PIN STATEMENT. HAVE FUN."""


# check for correct pin
def check_pin(pin):
    return pin == 1234


# Allow the user three tries to try the pin and validates correct pin
def log_in():
    attempts = 0
    while attempts < 3:
        four_digit_pin = int(input("Kindly Enter Your 4-Digit Pin: "))
        if check_pin(four_digit_pin):
            print("Pin approved.")
            return True
        else:
            print("Pin declined.")
            attempts += 1
    print("Too many incorrect attempts. unable to log in")
    return False


# Check if amount input is valid
def check_input(amount):
    try:
        # check if inserted amount is an integer
        withdrawal_amount = int(amount)
        try:
            # raise an error if withdrawal amount is negative
            if withdrawal_amount <= 0:
                raise ValueError("Please enter a valid amount.")
        except ValueError as v_error:
            print(v_error)
            return False
    except ValueError:
        print("Invalid input")
        return False
    else:
        return True


# Function to allow money withdrawal
def transaction(amount):
    # Set account balance to 100
    balance = 100
    try:
        equity = balance - amount
        # checking if balance is negative
        if equity < 0:
            raise ValueError("You do not have the sufficient funds to complete this transaction")
    except ValueError as v_error:
        print(v_error)
    else:
        # minus amount from total balance
        balance -= amount
        print(f"Withdrawal successful. Your remaining balance is: ${balance}")
    return balance


def atm():
    print("Welcome to Wofai's Python ATM")
    if log_in():
        amount = input('please enter withdrawal amount: $')
        if check_input(amount):
            transaction(int(amount))


atm()
```
