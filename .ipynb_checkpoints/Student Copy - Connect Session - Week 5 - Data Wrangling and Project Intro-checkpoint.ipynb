### Scrap table from website

```python
from bs4 import BeautifulSoup
```

```python
url ='https://en.wikipedia.org/wiki/World_population_milestones'

data = requests.get(url).text # save the page as texts
soup = BeautifulSoup(data, 'html.parser') # create the soup
table = soup.find_all('table') #find all tables
table
```

```python
len(table)
```

```python
table[3]
```

```python
import warnings
warnings.filterwarnings('ignore')
```

### [Table Structure example](https://www.freecodecamp.org/news/html-tables-table-tutorial-with-css-example-code/)

```python
# create a DataFrame from the html table

regional_pop = pd.DataFrame(columns = ['Country/Nationality','Population','Born(Date)','Name','Notes','Footnotes'])

for row in table[3].tbody.find_all('tr'):
    col = row.find_all('td')
    if (col != []):
        country = col[0].text
        population = col[1].text
        born = col[2].text
        name = col[3].text
        notes  = col[4].text
        footnotes = col[5].text
        regional_pop = regional_pop.append({'Country/Nationality':country, 
                                            'Population':population, 
                                            'Born(Date)':born,
                                            'Name':name,
                                            'Notes':notes,
                                            'Footnotes':footnotes
                                           }, ignore_index=True)
```

## Assessing Data

* Assessing is different from Exploring
* Assessing the data for quality and structure, makes the data work
* Exploring (EDA) takes you deeper to understand your data using simple visuals, e.g. removing outliers and create new * descriptive features i.e. feature engineering, summarize characteristics
* Sometimes they happen together

* Is our data complete? - Missing values
* Is our data valid? - inconsistency or wrong types
* Is our data accurate? - inaccurate values, strange figures, outliers
* Is our data consistent? - validity and accuracy issues

**Inspect, Detect and Document**

* Visual Assessment - note what we observe from visual inspection
* Programmatic Assessment - re-examine and reconfirm what we have noted down and check again for more issues and validate the specifics and quantities of those issues

**Assess for Quality Issues (Dirty data)**
* Missing data
* Duplicate data
* Incorrect data
* Corrupted Data
